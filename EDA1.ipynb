```python
### To replace null values we use the mean median and the mode,
```

```python
df['Price (INR)'] = df['Price (INR)'].fillna(df['Price (INR)'].mean())
```

```python
df['Fuel Type'] = df['Fuel Type'].fillna(df['Fuel Type'].mode()[0])
```

```python
# IQR

Q1 = df['Price (INR)'].quantile(0.25)
Q2 = df['Price (INR)'].quantile(0.75)

IQR = Q2 - Q1



outlier = df[(df['Price (INR)'] < (Q1 - 1.5*IQR))]
```

```python
offers = pd.DataFrame({
    'Car Brand' : ['Maruti' , 'Honda' , 'Hyundai'],
    'Discount' : [50000 , 30000 , 40000]
})
```
